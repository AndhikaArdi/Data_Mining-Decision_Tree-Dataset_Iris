```python
from sklearn.tree import DecisionTreeClassifier
import pandas as pd
import matplotlib.pyplot as plt
```

```python
iris = pd.read_csv('Dataset_Iris.csv')
target_mapping = {'setosa': 0, 'versicolor': 1, 'virginica': 2}
iris['species'] = iris['species'].replace(target_mapping)
features = iris.iloc[:, 0:4].values
target = iris.iloc[:, -1].values
```

```python
Decisiontree = DecisionTreeClassifier(
    random_state=0,
    max_depth=None,
    min_samples_split=2,
    min_samples_leaf=1,
    min_weight_fraction_leaf=0,
    max_leaf_nodes=None,
    min_impurity_decrease=0
)
```

```python
model = Decisiontree.fit(features, target)
```

```python
observation = [[5,4,3,2]]
model.predict(observation)
model.predict_proba(observation)
```

```python
import pydotplus
from sklearn import tree

data_names = iris.iloc[:, 0:4].columns.values.tolist()
target_names = ['setosa','versicolor','virginica']

dot_data = tree.export_graphviz(Decisiontree, out_file=None, feature_names = data_names, class_names = target_names)
from IPython.display import Image
graph = pydotplus.graph_from_dot_data(dot_data)
Image(graph.create_png())
graph.write_png("D:\iris.png")
```
